# Trisha — Mid-Size Models
RTX 3050 6GB | batch_size=32 | accumulation_steps=1

```python
import sys
sys.path.insert(0, "../..")

from modules.utils.config import *
from modules.utils.seed import set_seed
from modules.data.dataset import get_dataloaders, get_test_loader
from modules.models.factory import TrashClassifier
from modules.training.train import fit

set_seed(SEED)
print(f"Device: {DEVICE}")
```

```python
# MODELS = [
#     "resnet18",
#     "resnet34",
#     "densenet121",
#     "efficientnet_b1",
#     "efficientnet_b2",
# ]

BATCH_SIZE = 32
ACCUMULATION_STEPS = 1
NUM_WORKERS = 2
EPOCHS_HEAD = 10
EPOCHS_FINETUNE = 20
LR_HEAD = 1e-3
LR_FINETUNE = 1e-4
PATIENCE = 10
```

```python
train_loader, val_loader, val_ds = get_dataloaders(
    batch_size=BATCH_SIZE,
    num_workers=NUM_WORKERS,
)
print(f"Train: {len(train_loader.dataset)} | Val: {len(val_loader.dataset)}")
```

```python
results = []
```

## **ResNet18**

```python
model = TrashClassifier("resnet18", num_classes=NUM_CLASSES).to(DEVICE)
res = fit(
    model, train_loader, val_loader,
    name="trisha_resnet18",
    encoder_name="resnet18",
    accumulation_steps=ACCUMULATION_STEPS,
    epochs_head=EPOCHS_HEAD,
    epochs_finetune=EPOCHS_FINETUNE,
    lr_head=LR_HEAD,
    lr_finetune=LR_FINETUNE,
    patience=PATIENCE,
    class_weights=val_ds.class_weights,
    device=DEVICE,
)
results.append(res)
print(f">>> resnet18: val_f1_macro = {res['best_val_f1']:.4f}")
```

## **Ensemble + Submission**

```python
import json
import pandas as pd

records = []
for f in sorted(RESULTS.glob("*.json")):
    with open(f) as fh:
        records.append(json.load(fh))

summary = pd.DataFrame(records)
summary = summary.sort_values("best_val_f1", ascending=False)
summary[["name", "best_val_f1", "f1_per_class"]].to_csv(RESULTS / "trisha_summary.csv", index=False)
summary[["name", "best_val_f1", "f1_per_class"]]
```

```python
from modules.ensemble.vote import soft_voting, predict_ensemble, generate_submission

models, records = soft_voting(num_models=len(list(RESULTS.glob("*.json"))), device=DEVICE)
print("Models used for ensemble:")
for r in records:
    print(f"  {r['name']}: f1={r['best_val_f1']:.4f}")

BATCH_SIZE_TEST = 64
test_loader = get_test_loader(batch_size=BATCH_SIZE_TEST, num_workers=NUM_WORKERS)

pred_probs = predict_ensemble(models, test_loader, device=DEVICE)
generate_submission(pred_probs, output_path=RESULTS / "trisha_submission.csv")
```
