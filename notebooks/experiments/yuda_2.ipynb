# Yuda 2 — Batch Experiments
4 experiments sequential: Swin-T, EfficientNet-B3 @300, ConvNeXt+RandAugment, ConvNeXt+FocalLoss

```python
import sys
sys.path.insert(0, "../..")

import torch
import torch.nn as nn
import pandas as pd

from modules.utils.config import *
from modules.utils.seed import set_seed
from modules.data.dataset import get_dataloaders
from modules.models.factory import TrashClassifier
from modules.training.train import fit

set_seed(SEED)
print(f"Device: {DEVICE}")
```

```python
NUM_WORKERS = 4
EPOCHS_HEAD = 10
EPOCHS_FINETUNE = 20
LR_HEAD = 1e-3
LR_FINETUNE = 1e-4
PATIENCE = 10
ACCUMULATION_STEPS = 1

class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce = nn.functional.cross_entropy(inputs, targets, weight=self.alpha, reduction="none")
        pt = (-ce).exp()
        return (pt.clamp(min=1e-8).pow(self.gamma) * ce).mean()
```

## Experiment 1: Swin-T (Efficient Transformer)

```python
print("\n" + "=" * 50)
print("Experiment 1: Swin-T")
print("=" * 50)

train_loader, val_loader, val_ds = get_dataloaders(batch_size=32, num_workers=NUM_WORKERS)

model = TrashClassifier("swin_tiny_patch4_window7_224", num_classes=NUM_CLASSES)
res = fit(
    model, train_loader, val_loader,
    name="yuda2_swin_tiny",
    encoder_name="swin_tiny_patch4_window7_224",
    accumulation_steps=ACCUMULATION_STEPS,
    epochs_head=EPOCHS_HEAD,
    epochs_finetune=EPOCHS_FINETUNE,
    lr_head=LR_HEAD,
    lr_finetune=LR_FINETUNE,
    patience=PATIENCE,
    class_weights=val_ds.class_weights,
    device=DEVICE,
)
results = [res]
```

## Experiment 2: EfficientNet-B3 @ 300px

```python
print("\n" + "=" * 50)
print("Experiment 2: EfficientNet-B3 @ 300px")
print("=" * 50)

train_loader, val_loader, val_ds = get_dataloaders(batch_size=16, num_workers=NUM_WORKERS, img_size=300)

model = TrashClassifier("efficientnet_b3", num_classes=NUM_CLASSES)
res = fit(
    model, train_loader, val_loader,
    name="yuda2_effnet_b3_300",
    encoder_name="efficientnet_b3",
    accumulation_steps=ACCUMULATION_STEPS,
    epochs_head=EPOCHS_HEAD,
    epochs_finetune=EPOCHS_FINETUNE,
    lr_head=LR_HEAD,
    lr_finetune=LR_FINETUNE,
    patience=PATIENCE,
    class_weights=val_ds.class_weights,
    device=DEVICE,
)
results.append(res)
```

## Experiment 3: ConvNeXt + RandAugment

```python
print("\n" + "=" * 50)
print("Experiment 3: ConvNeXt + RandAugment")
print("=" * 50)

train_loader, val_loader, val_ds = get_dataloaders(batch_size=32, num_workers=NUM_WORKERS, rand_augment=True)

model = TrashClassifier("convnext_tiny", num_classes=NUM_CLASSES)
res = fit(
    model, train_loader, val_loader,
    name="yuda2_convnext_randaug",
    encoder_name="convnext_tiny",
    accumulation_steps=ACCUMULATION_STEPS,
    epochs_head=EPOCHS_HEAD,
    epochs_finetune=EPOCHS_FINETUNE,
    lr_head=LR_HEAD,
    lr_finetune=LR_FINETUNE,
    patience=PATIENCE,
    class_weights=val_ds.class_weights,
    device=DEVICE,
)
results.append(res)
```

## Experiment 4: ConvNeXt + Focal Loss

```python
print("\n" + "=" * 50)
print("Experiment 4: ConvNeXt + Focal Loss")
print("=" * 50)

train_loader, val_loader, val_ds = get_dataloaders(batch_size=32, num_workers=NUM_WORKERS)

model = TrashClassifier("convnext_tiny", num_classes=NUM_CLASSES)
focal_loss = FocalLoss(alpha=val_ds.class_weights.to(DEVICE), gamma=2.0)
res = fit(
    model, train_loader, val_loader,
    name="yuda2_convnext_focal",
    encoder_name="convnext_tiny",
    accumulation_steps=ACCUMULATION_STEPS,
    epochs_head=EPOCHS_HEAD,
    epochs_finetune=EPOCHS_FINETUNE,
    lr_head=LR_HEAD,
    lr_finetune=LR_FINETUNE,
    patience=PATIENCE,
    criterion=focal_loss,
    device=DEVICE,
)
results.append(res)
```

## Summary

```python
summary = pd.DataFrame(results)
summary = summary.sort_values("best_val_f1", ascending=False)
summary.to_csv(RESULTS / "yuda2_summary.csv", index=False)
summary[["name", "best_val_f1", "f1_per_class"]]
```
